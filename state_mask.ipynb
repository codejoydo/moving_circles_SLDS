```python
import pickle
# from glob import glob
from os.path import exists

K = 6
D = 10
N = 85
num_subjs = 92
M = 20
num_resamples = 100
```

```python
with open(f"pkl/state_masks_K{K}_D{D}_N{N}_{num_subjs}subjs_{num_resamples}resamples.pkl", "rb") as f:
    state_masks = pickle.load(f)
```

```python
for idx_state in range(K):
    files = []
    for idx_resample in range(1,1+num_resamples):
        mask = state_masks[idx_resample][idx_state]
        if mask==0:
            fname = f"/media/joyneelm/joy_backup/slds/results/group_level/rslds_K6_D10_N85_92subjs_emoprox_resample{idx_resample}/state{idx_state+1}.nii.gz"
            if exists(fname):
                files.append(fname)
    
    print(len(files))
    print(" ".join(files))

    with open(f"/media/joyneelm/joy_backup/slds/state{idx_state+1}_files.txt", "w") as f:
        f.write(" ".join(files))
```
